```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv('/kaggle/input/german-credit/german_credit_data.csv')
df.head()
```

```python
print('Info\n', df.info())
print('Describe\n', df.describe())
```

```python
print('Sex:\n', df['Sex'].unique())
print('Job:\n', df['Job'].unique())
print('Housing:\n', df['Housing'].unique())
print('Saving accounts:\n', df['Saving accounts'].unique())
print('Checking account:\n', df['Checking account'].unique())
print('Purpose:\n', df['Purpose'].unique())
```

```python
df = df.dropna()
```

```python
%matplotlib inline
df.hist(bins=50,figsize=(20,15))
```

```python
plt.figure(figsize=(25, 10))
sns.boxplot(x="Job", y="Credit amount", data=df)
```

```python
sns.heatmap(df.corr(),annot=True,linewidths=.5)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()
df['Sex'] = labelencoder_X.fit_transform(df['Sex'])
df['Housing'] = labelencoder_X.fit_transform(df['Housing']) 
df['Saving accounts'] = labelencoder_X.fit_transform(df['Saving accounts']) 
df['Checking account'] = labelencoder_X.fit_transform(df['Checking account']) 
df['Purpose'] = labelencoder_X.fit_transform(df['Purpose']) 
```
